1. 基本文本处理技能
1.1 分词的概念及方法
Word segmentation：the problem of dividing a string of written language into its component words.
常用的字符串匹配方法：
    正向最大匹配法（从左到右的方向）；
    逆向最大匹配法（从右到左的方向）；
    最小切分（每一句中切出的词数最小）；
    双向最大匹配（进行从左到右、从右到左两次扫描）。
Jieba分词安装：conda install -c conda-forge jieba直接安装（mac/linux在终端输入指令；Windows在Anaconda Prompt 输入指令）。
1.2 词、字符频率统计（Python中的collections.Counter模块）
    基于统计的分词方法是在给定大量已经分词的文本的前提下，利用统计机器学习模型学习词语切分的规律（称为训练），从而实现对未知文本的切分。
    主要的统计模型有：N元文法模型（N-gram），隐马尔可夫模型（Hidden Markov Model ，HMM），最大熵模型（ME），条件随机场模型（Conditional Random Fields，CRF）等。
    基于统计的分词方法包括：N-最短路径方法、基于词的n元语法模型的分词方法、由字构词的汉语分词方法、基于词感知机算法的汉语分词方法、基于字的生成式模型和区分式模型相结合的汉语分词方法。
2. 语言模型中unigram、bigram、trigram的概念；
unigram 一元分词，把句子分成一个一个的汉字
bigram 二元分词，把句子从头到尾每两个字组成一个词语
trigram 三元分词，把句子从头到尾每三个字组成一个词语.

3. 文本矩阵化：要求采用词袋模型且是词级别的矩阵化
步骤： 分词\去停用词 构造词表\向量化。


```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
from collections import Counter
import jieba

def get_text(text_path):
    with open(text_path,'r',encoding='utf8') as f:
        text= f.readlines()
    return text

text_list = get_text('C:/Users/a/py files/cnew_train.txt')
line = text_list[0]
print(line)
```

```python
line = line.strip('\t')
words_list = list(jieba.cut(line, cut_all=False))
print(words_list)
```

```python
# 统计词频
words_frequency=dict()
for word in words_list:
    if word not in words_frequency:
        words_frequency[word]=1
    else:
        words_frequency[word]+=1
print(words_frequency)

# 也可用Counter
count_character = Counter(line)
print(count_character)
count_words = Counter(words_list)
print(count_words)
```
